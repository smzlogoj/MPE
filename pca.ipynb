```r
library(factoextra)
library(FactoMineR)
library(ggpubr)
```

```r
res.pca <- PCA(df[, 6:76], scale.unit = TRUE, ncp = 15)
```

```r
fviz_eig(res.pca, addlabels = TRUE, ylim = c(0, 30))
```

```r
p1 <- ggplot(as.data.frame(res.pca$ind$coord), aes(x = Dim.1, y = Dim.2)) +
        geom_point(aes(color = df$class), shape = 19, size = 1) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        labs(color = 'Clases') +
        theme_bw()
p2 <- ggplot(as.data.frame(res.pca$ind$coord), aes(x = Dim.1, y = Dim.3)) +
        geom_point(aes(color = df$class), shape = 19, size = 1) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw()
p3 <- ggplot(as.data.frame(res.pca$ind$coord), aes(x = Dim.2, y = Dim.3)) +
        geom_point(aes(color = df$class), shape = 19, size = 1) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw()
p4 <- ggplot(as.data.frame(res.pca$ind$coord), aes(x = Dim.2, y = Dim.4)) +
        geom_point(aes(color = df$class), shape = 19, size = 1) +
        geom_hline(yintercept = 0, linetype = "solid", color = "black", size = .2) +
        geom_vline(xintercept = 0, linetype = "solid", color = "black", size = .2) +
        theme_bw()

ggarrange(p1, p2, p3, p4, 
            nrow = 2, ncol = 2, common.legend  = TRUE, legend = 'bottom', align = 'hv')
```
